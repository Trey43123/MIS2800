```python
%%sql
CREATE TABLE Marketing_Brands (
  brand_id INTEGER PRIMARY KEY,
  brand_name TEXT,
  brand_type TEXT,
  brand_budget INTEGER
) STRICT;
```

```python
%%sql
-- Clear the table first so we can rerun this cell without unique constraint errors
DELETE FROM Marketing_Brands;

-- Fixed: Added 'INTO Marketing_Brands' and removed underscores from the brand names
INSERT INTO Marketing_Brands (brand_id, brand_name, brand_type, brand_budget)
VALUES
  (1, 'Doritos', 'Snacks', 7000000),
  (2, 'AWS', 'Technology', 15000000),
  (3, 'Pepsi', 'Beverages', 2000000),
  (4, 'He Gets Us', 'Religious', 1000000),
  (5, 'Dr Pepper', 'Beverages', 5000000),
  (6, 'IBM', 'Technology', 100000),
  (7, 'Google', 'Technology', 8000000);

-- Query the table to verify the data
SELECT * FROM Marketing_Brands;
```

```python
%%sql
SELECT * FROM Marketing_Brands;
```

```python
%%sql
-- Fixed: Selected specific columns and added a filter for budget > 700,000
SELECT brand_name, brand_type
FROM Marketing_Brands
WHERE brand_budget > 700000;
```

```python
%%sql
-- Fixed: Added column selection and a WHERE clause to filter for 'Religious' brand types
SELECT brand_name, brand_type
FROM Marketing_Brands
WHERE brand_type = 'Religious';
```

```python
%%sql
-- Fixed: Selected brand_name and added a filter for budget < 1,000,000
SELECT brand_name
FROM Marketing_Brands
WHERE brand_budget < 1000000;
```

```python
%%sql
SELECT * FROM Marketing_Brands WHERE brand_budget > 7000000
```

```python
%%sql
-- Block 8
-- Find the brands where the brand type is Snacks or Beverages. Display only the brand names.
SELECT brand_name
FROM Marketing_Brands
WHERE brand_type IN ('Snacks', 'Beverages');
```

```python
%%sql
-- Block 9
-- Display only the brand_name and brand_budget, sorted by budget in descending order.
SELECT brand_name, brand_budget
FROM Marketing_Brands
ORDER BY brand_budget DESC;
```

```python
%%sql
-- Block 10
-- Show all columns, sorted by budget in ascending order
SELECT * FROM Marketing_Brands
ORDER BY brand_budget ASC;
```
